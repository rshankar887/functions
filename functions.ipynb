3. What are the different ways to define and call a function in Python?
  - **1. Basic Function Definition and Call:**

      This is the most common way.  You define a function using the `def` keyword, followed by the function name, parentheses `()`, and a colon `:`. The code block within the function is indented.

```python
# Function definition
def greet(name):
    """Greets the person passed in as a parameter."""  # Docstring (optional but recommended)
    print(f"Hello, {name}!")

# Function call
greet("Alice")  # Output: Hello, Alice!
greet("Bob")    # Output: Hello, Bob!
```

**2. Functions with Default Parameter Values:**

You can provide default values for parameters. If an argument is not provided for a parameter with a default value during the function call, the default value is used.

```python
def power(base, exponent=2):
    return base ** exponent

print(power(3))     # Output: 9 (exponent defaults to 2)
print(power(3, 3))  # Output: 27 (exponent is overridden)
```

**3. Functions with Variable-Length Arguments (`*args`):**

The `*args` syntax allows a function to accept any number of positional arguments. These arguments are collected into a tuple.

```python
def print_numbers(*args):
    for num in args:
        print(num)

print_numbers(1, 2, 3, 4, 5)  # Output: 1 2 3 4 5 (each on a new line)
```

**4. Functions with Variable-Length Keyword Arguments (`**kwargs`):**

The `**kwargs` syntax allows a function to accept any number of keyword arguments. These arguments are collected into a dictionary.

```python
def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="Alice", age=30, city="New York")  # Output: name: Alice, age: 30, city: New York (each on a new line)
```

**5. Functions with Both `*args` and `**kwargs`:**

You can combine both to accept any number of positional and keyword arguments.  `*args` must come before `**kwargs` in the function definition.

```python
def print_all(a, b, *args, **kwargs):
    print(f"a: {a}, b: {b}")
    print("Positional arguments:", args)
    print("Keyword arguments:", kwargs)

print_all(1, 2, 3, 4, name="Alice", age=30)
# Output:
# a: 1, b: 2
# Positional arguments: (3, 4)
# Keyword arguments: {'name': 'Alice', 'age': 30}
```

**6. Lambda Functions (Anonymous Functions):**

Lambda functions are small, anonymous functions defined using the `lambda` keyword. They can have any number of arguments but can only contain one expression.  They are often used for short, simple operations.

```python
square = lambda x: x * x
print(square(5))  # Output: 25

add = lambda x, y: x + y
print(add(2, 3))   # Output: 5
```

**7. Calling Functions with Different Argument Types:**

As shown in the examples above, you can call functions using:

* **Positional arguments:**  `greet("Alice")`
* **Keyword arguments:** `greet(name="Alice")`
* **A combination:** `print_all(1, 2, 3, name="Alice")`
* **Unpacking sequences (lists, tuples) for positional arguments:**
  ```python
  args = (1, 2, 3)
  print_numbers(*args)  # Equivalent to print_numbers(1, 2, 3)
  ```
* **Unpacking dictionaries for keyword arguments:**

**8. Recursive Functions:**

A recursive function is a function that calls itself.  It's essential to have a base case to stop the recursion and prevent infinite loops.

```python
def factorial(n):
    if n == 0:  # Base case
        return 1
    else:
        return n * factorial(n-1)  # Recursive call

print(factorial(5))  # Output: 120
```

These are the primary ways to define and call functions in Python.  Understanding these different methods provides you with the flexibility to write more modular, reusable, and efficient code.  Remember to choose the approach that best suits the specific task you're trying to accomplish.


4. What is the purpose of the `return` statement in a Python function?
    - The `return` statement in a Python function serves several crucial purposes:

1. **Returning a Value:** The most common purpose is to send a value back to the caller of the function.  This value can be of any data type (integer, float, string, list, tuple, dictionary, object, etc.).  If a `return` statement is encountered, the function's execution immediately stops, and the specified value is returned.

   ```python
   def add(x, y):
       return x + y  # Returns the sum of x and y

   result = add(5, 3)  # The returned value (8) is assigned to 'result'
   print(result)       # Output: 8
   ```

2. **Exiting the Function:** Even if a function doesn't need to return a specific value, the `return` statement can be used to exit the function's execution at a particular point.  This is useful for situations where you want to stop the function's execution based on a certain condition.  If no value is specified after `return`, the function implicitly returns `None`.

   ```python
   def check_value(x):
       if x < 0:
           print("Value is negative. Exiting.")
           return  # Exits the function if x is negative
       print("Value is positive or zero.")

   check_value(-5)  # Output: Value is negative. Exiting.
   check_value(5)   # Output: Value is positive or zero.
   ```

3. **Returning Multiple Values:** You can return multiple values from a function by packing them into a tuple or a list.  The caller can then unpack these values.

   ```python
   def get_coordinates():
       x = 10
       y = 20
       return x, y  # Returns x and y as a tuple

   x_coord, y_coord = get_coordinates()  # Unpacking the returned tuple
   print(f"x: {x_coord}, y: {y_coord}")  # Output: x: 10, y: 20

   # Or return a dictionary for more clarity:
   def get_user_info():
       return {"name": "Alice", "age": 30}

   user = get_user_info()
   print(user["name"])  # Output: Alice
   ```

4. **Control Flow:**  `return` statements play a role in controlling the flow of execution within a program.  They allow you to define how and when a function should terminate and what data it should provide back to the part of the code that called it.

**In summary:**

* `return value`: Returns a specific value to the caller.
* `return`: Exits the function (implicitly returns `None`).
* `return value1, value2, ...`: Returns multiple values as a tuple.

Understanding the `return` statement is fundamental to writing effective and reusable functions in Python. It's how functions communicate results and control the flow of your program.


5. What are iterators in Python and how do they differ from iterables?
    - In Python, both iterables and iterators are related concepts that deal with sequences of data, but they have distinct roles:

**Iterables:**

* **Definition:** An iterable is any object in Python that can return its members one at a time.  Essentially, it's something you can loop over using a `for` loop.
* **Examples:** Lists, tuples, strings, dictionaries, sets, files, and user-defined objects that implement the *iterable protocol* are all iterables.
* **How they work:** Iterables provide data on demand. They don't load everything into memory at once, which is very efficient for large datasets.
* **Key method:** Iterables implement the `__iter__()` method. This method returns an *iterator* object.

**Iterators:**

* **Definition:** An iterator is an object that produces a sequence of values one at a time.  It has a state so it knows where it is in the sequence.
* **How they work:** Iterators produce the next value in the sequence when you ask for it.  Once it has returned all values, it raises a `StopIteration` exception to signal the end of the sequence.
* **Key methods:** Iterators implement two methods:
    * `__iter__()`: Returns the iterator object itself (which is often the case).
    * `__next__()`: Returns the next value in the sequence.  Raises `StopIteration` when there are no more items.

**Key Differences Summarized:**

| Feature        | Iterable                                    | Iterator                                      |
|----------------|---------------------------------------------|---------------------------------------------|
| **Purpose**    | Provides data on demand                      | Produces values one at a time                |
| **`__iter__()`** | Returns an iterator                         | Returns itself                               |
| **`__next__()`** | Not implemented directly                     | Implemented to get the next value; raises `StopIteration` |
| **Usability**  | Can be used in a `for` loop directly        | Used by iterables and implicitly by `for` loops |
| **State**      | Doesn't maintain state about iteration       | Maintains state (current position)            |

**Analogy:**

Think of a library with many books.

* The **library** itself is an *iterable*.  It contains a collection of books. You can go there and browse.
* A **librarian** is like an *iterator*. They can fetch you the next book in a specific order when you ask for it. They remember where they are in the collection.

**Example:**

```python
# An iterable (a list)
my_list = [1, 2, 3]

# Get an iterator from the iterable
my_iterator = iter(my_list)  # Calls my_list.__iter__()

# Get the next value from the iterator
print(next(my_iterator))  # Output: 1
print(next(my_iterator))  # Output: 2
print(next(my_iterator))  # Output: 3

# Trying to get the next value after the end raises StopIteration
try:
    print(next(my_iterator))  # Raises StopIteration
except StopIteration:
    print("End of iteration")

# Looping through an iterable (implicitly uses an iterator)
for item in my_list:
    print(item)  # Output: 1 2 3 (each on a new line)
```

**In essence:** Iterables are the source of data, and iterators are the mechanism to access that data sequentially.  The `for` loop in Python handles the details of getting an iterator from an iterable and repeatedly calling `next()` until `StopIteration` is raised. This makes working with sequences in Python efficient and memory-friendly, especially for large amounts of data.


6. Explain the concept of generators in Python and how they are defined.
    - Generators in Python are a special type of iterator. They are a powerful and memory-efficient way to create sequences of values on the fly, rather than storing them all in memory at once.  This is particularly useful when dealing with large datasets or infinite sequences.

**Key characteristics of generators:**

1. **Iterator:** Generators are iterators, meaning you can use them in `for` loops and with functions like `next()`.

2. **Lazy Evaluation:** Generators produce values on demand (lazy evaluation). They don't generate all values at once; they generate each value only when it's requested. This saves memory.

3. **Defined using functions:** Generators are defined like regular functions but use the `yield` keyword instead of `return`.

4. **Stateful:** Generators maintain their internal state between calls.  When you call `next()` on a generator, it resumes execution from where it left off, remembers the values of its variables, and yields the next value.

**How generators are defined:**

Generators are defined using a function-like syntax, but with the `yield` keyword.  Here's a simple example:

```python
def my_generator():
    yield 1
    yield 2
    yield 3

# Create a generator object
gen = my_generator()

# Get the next value
print(next(gen))  # Output: 1
print(next(gen))  # Output: 2
print(next(gen))  # Output: 3

# When there are no more values, StopIteration is raised
try:
    print(next(gen))  # Raises StopIteration
except StopIteration:
    print("End of generation")

# You can also use a for loop with a generator
for value in my_generator():
    print(value)  # Output: 1 2 3 (each on a new line)
```

**How `yield` works:**

* When a generator function encounters a `yield` statement, it pauses execution and "yields" the specified value to the caller.
* The generator's state is saved, so when the caller requests the next value, the generator resumes execution from where it left off (after the `yield` statement).
* The generator continues to run until it encounters another `yield` or until the function finishes.  If the function finishes, a `StopIteration` exception is automatically raised.

**Example with a loop:**

```python
def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1

for num in count_up_to(5):
    print(num)  # Output: 1 2 3 4 5 (each on a new line)
```

**Generator expressions:**

Just like list comprehensions, there are also generator expressions, which provide a concise way to create generators:

```python
squares = (x * x for x in range(10))  # Creates a generator of squares

for square in squares:
    print(square)  # Output: 0 1 4 9 16 ... 81 (each on a new line)
```

**Benefits of using generators:**

* **Memory efficiency:**  Generators save memory by generating values on demand.  They don't store the entire sequence in memory.
* **Performance:**  Lazy evaluation can improve performance, especially when you don't need to process all the values in the sequence.
* **Readability:**  Generators can make code more readable and easier to understand, especially when dealing with complex sequences.
* **Infinite sequences:** You can create generators that produce infinite sequences of values, as they don't need to be stored in memory.




7. What are the advantages of using generators over regular functions?
    - Generators offer several significant advantages over regular functions when dealing with sequences of data:

1. **Memory Efficiency:** This is the most prominent advantage. Generators produce values on demand (lazy evaluation).  They don't store the entire sequence in memory at once.  Instead, they generate each value only when it's requested. This is crucial when working with large datasets that would otherwise consume excessive memory if stored entirely.  Regular functions, on the other hand, typically create the entire sequence in memory before returning it.

2. **Performance (for certain use cases):** Because generators produce values on demand, they can be more performant if you don't need to process all the values in the sequence.  For example, if you're searching for a specific item in a large dataset, a generator can stop as soon as it finds the item, without having to generate the rest of the sequence.  Regular functions would have to generate the entire sequence first.  However, for operations that require processing the entire sequence, generators might have a slight performance overhead due to the context switching involved in yielding values.

3. **Readability and Code Clarity:** Generators can make code more readable and easier to understand, especially when dealing with complex sequence generation logic.  The `yield` keyword allows you to express the sequence generation process in a more natural and concise way, similar to how you would describe it step by step.

4. **Handling Infinite Sequences:**  You can create generators that produce infinite sequences of values.  Since they generate values on demand, they don't need to store the entire infinite sequence in memory.  This is impossible with regular functions that create the entire sequence at once.

5. **Simplified Code for Sequence Operations:** Generators can simplify code for many sequence-related operations, such as filtering, mapping, and reducing.  You can chain generators together to create complex data processing pipelines in a very elegant and efficient way.

**Example illustrating memory efficiency:**

```python
# Regular function (creates the whole list in memory)
def regular_function(n):
    result = []
    for i in range(n):
        result.append(i * 2)
    return result

# Generator function (generates values on demand)
def generator_function(n):
    for i in range(n):
        yield i * 2

# For large values of n, the generator is much more memory-efficient
n = 10**9  # A very large number

# Using the regular function would likely lead to a MemoryError
# large_list = regular_function(n)  # Don't do this!

# The generator is perfectly fine, as it generates values one at a time
large_generator = generator_function(n)

# Process the generator without loading everything into memory
for value in large_generator:
    # Do something with the value
    if value > 1000:  # Example: stop processing early
        break

```

**When to use regular functions:**

While generators are often preferred for sequence generation, regular functions are still appropriate in many situations:

* When you need to create a small, fixed-size sequence and store it entirely in memory.
* When the logic for generating the sequence is very simple.
* When you need to perform operations that require the entire sequence to be available at once.
* When performance is absolutely critical and the overhead of yielding values becomes a bottleneck (though this is less common).

In most cases where you're working with sequences, especially large ones, generators offer a more efficient, readable, and elegant solution.  They are a key feature of Python that promotes memory-conscious and high-performance programming.


8. What is a lambda function in Python and when is it typically used?
    - A lambda function in Python is a small, anonymous, and unnamed function defined using the `lambda` keyword.  It can take any number of arguments but can only contain a single expression. The result of this expression is implicitly returned by the lambda function.

**Syntax:**

```python
lambda arguments: expression
```

* `lambda`: The keyword that indicates you're defining a lambda function.
* `arguments`: A comma-separated list of input parameters (can be zero or more).
* `expression`: A single expression that is evaluated and returned.

**Example:**

```python
square = lambda x: x * x
print(square(5))  # Output: 25

add = lambda x, y: x + y
print(add(2, 3))  # Output: 5

greet = lambda name: f"Hello, {name}!"
print(greet("Alice"))  # Output: Hello, Alice!
```

**When are lambda functions typically used?**

Lambda functions are best suited for short, simple operations where a full function definition would be overly verbose. They are often used in the following situations:

1. **As arguments to higher-order functions:**  Higher-order functions are functions that take other functions as arguments. Lambda functions are frequently used to provide simple functions as arguments to functions like `map()`, `filter()`, `sorted()`, and `reduce()`.

   ```python
   numbers = [1, 2, 3, 4, 5]

   # Using map() with a lambda function to square each number
   squared_numbers = list(map(lambda x: x * x, numbers))
   print(squared_numbers)  # Output: [1, 4, 9, 16, 25]

   # Using filter() with a lambda function to get even numbers
   even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
   print(even_numbers)  # Output: [2, 4]

   # Using sorted() with a lambda function to sort by the second element of tuples
   data = [(1, 5), (3, 2), (2, 8)]
   sorted_data = sorted(data, key=lambda item: item[1])
   print(sorted_data)  # Output: [(3, 2), (1, 5), (2, 8)]
   ```

2. **For creating simple, one-time-use functions:** If you need a function for a very specific purpose and you don't want to define a named function, a lambda function can be a quick and convenient solution.

3. **Within other functions (nested functions):**  Sometimes, it's useful to define a small helper function inside another function.  Lambda functions can be used for this purpose, avoiding the need for a separate named function definition.

**When *not* to use lambda functions:**

* **Complex logic:** If the function's logic is more than a single expression, it's generally better to define a regular named function for better readability and maintainability. Lambda functions are not designed for multi-line statements or complex control flow.
* **Code reusability:** If you need to reuse the function in multiple places, define a regular named function. Lambda functions are anonymous and not easily reusable.
* **Documentation:** Named functions can have docstrings to explain their purpose and behavior. Lambda functions cannot have docstrings, making it more difficult to document their functionality.

**In summary:** Lambda functions are a handy tool for creating small, simple functions on the fly. They are particularly useful when working with higher-order functions or when you need a function for a one-time use. However, for more complex logic or when reusability and documentation are important, regular named functions are the better choice.


9. Explain the purpose and usage of the `map()` function in Python.
    -The `map()` function in Python is a built-in function that applies a given function to each item of an iterable (like a list, tuple, etc.) and returns an iterator that yields the results.  It's a concise way to perform the same operation on a sequence of elements.

**Purpose:**

The primary purpose of `map()` is to transform each element of an iterable using a specific function, creating a new iterable with the transformed elements.  It avoids the need for explicit loops in many cases, making the code more readable and often more efficient.

**Usage:**

The `map()` function takes two arguments:

1. **`function`:** The function to apply to each item of the iterable.  This can be a regular function or a lambda function.
2. **`iterable`:** The iterable (e.g., list, tuple, set, string) whose elements will be processed.

**Syntax:**

```python
map(function, iterable, ...) # The ... signifies that you can add more iterables
```

**Return value:**

`map()` returns a *map object*, which is an iterator.  To get a list of the results, you typically need to convert the map object to a list using `list()`:

```python
result_list = list(map(function, iterable))
```

**Examples:**

1. **Squaring numbers in a list:**

   ```python
   numbers = [1, 2, 3, 4, 5]

   # Using a regular function
   def square(x):
       return x * x

   squared_numbers = list(map(square, numbers))
   print(squared_numbers)  # Output: [1, 4, 9, 16, 25]

   # Using a lambda function (more concise)
   squared_numbers = list(map(lambda x: x * x, numbers))
   print(squared_numbers)  # Output: [1, 4, 9, 16, 25]
   ```

2. **Converting strings to uppercase:**

   ```python
   strings = ["hello", "world"]
   uppercase_strings = list(map(str.upper, strings))  # str.upper is the function
   print(uppercase_strings)  # Output: ['HELLO', 'WORLD']
   ```

3. **Applying a function to multiple iterables:**

   `map()` can also be used with multiple iterables.  In this case, the function must take as many arguments as there are iterables.  The `map()` function will stop when the shortest iterable is exhausted.

   ```python
   list1 = [1, 2, 3]
   list2 = [4, 5, 6]

   result = list(map(lambda x, y: x + y, list1, list2))
   print(result)  # Output: [5, 7, 9]
   ```

**Advantages of using `map()`:**

* **Conciseness:**  `map()` often provides a more compact and readable way to perform operations on sequences compared to explicit loops, especially for simple transformations.
* **Functional approach:**  `map()` encourages a functional programming style, which can make code easier to reason about and less prone to errors.

**When not to use `map()`:**

* **Complex transformations:**  If the transformation logic is complex or involves multiple steps, a regular loop might be more readable and easier to maintain.
* **Conditional operations:**  If you need to apply the function conditionally based on certain criteria, a loop with `if` statements might be more appropriate.
* **Readability:**  While `map()` can be concise, if the lambda function or the function being used is too complex, it can sometimes make the code harder to read.  In such cases, a regular loop or list comprehension might be clearer.

**Alternatives to `map()`:**

* **List comprehensions:** List comprehensions often provide a more Pythonic and readable alternative to `map()` for simple transformations.
* **Generator expressions:** If you're working with large datasets, generator expressions can be more memory-efficient than `map()` as they produce values on demand.

In most cases, `map()` provides a convenient and efficient way to apply a function to a sequence of elements. However, it's important to choose the most appropriate approach based on the complexity of the transformation and the specific requirements of your code.


10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
    - `map()`, `reduce()`, and `filter()` are all built-in functions in Python that operate on sequences (iterables) of data. They are often used in functional programming paradigms. Here's a breakdown of their differences:

**1. `map()`:**

* **Purpose:** Applies a given function to each item of an iterable and returns an iterator that yields the results.  Transforms each element.
* **Output:** An iterator (map object) containing the transformed elements. You often need to convert it to a list using `list(map(...))` to see the results.
* **Example:**

```python
numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x * x, numbers))  # Squares each number
print(squared_numbers)  # Output: [1, 4, 9, 16, 25]
```

**2. `filter()`:**

* **Purpose:** Filters an iterable, keeping only the items for which a given function returns `True`.  Selects elements based on a condition.
* **Output:** An iterator (filter object) containing the filtered elements.  You often need to convert it to a list using `list(filter(...))` to see the results.
* **Example:**

```python
numbers = [1, 2, 3, 4, 5]
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))  # Keeps only even numbers
print(even_numbers)  # Output: [2, 4]
```

**3. `reduce()`:**

* **Purpose:** Applies a given function cumulatively to the items of a sequence, reducing it to a single value.  Combines elements using a binary operation.
* **Output:** A single value (not an iterator).
* **Note:** `reduce()` is part of the `functools` module in Python 3, so you need to import it: `from functools import reduce`
* **Example:**

```python
from functools import reduce

numbers = [1, 2, 3, 4, 5]
product = reduce(lambda x, y: x * y, numbers)  # Calculates the product of all numbers
print(product)  # Output: 120
```

**Key Differences Summarized:**

| Function  | Purpose                                      | Output                                   | Example                                     |
|-----------|----------------------------------------------|-------------------------------------------|---------------------------------------------|
| `map()`    | Transforms each element of an iterable        | Iterator (map object)                     | Squares each number in a list             |
| `filter()` | Selects elements from an iterable based on a condition | Iterator (filter object)                  | Gets even numbers from a list              |
| `reduce()` | Reduces a sequence to a single value         | Single value                               | Calculates the product of numbers in a list |

**Analogy:**

Imagine you have a basket of fruits:

* `map()`: You wash and polish each fruit (transforming them).
* `filter()`: You pick out only the ripe fruits (selecting based on a condition).
* `reduce()`: You blend all the fruits together into a smoothie (combining them into a single result).

**In essence:**

* `map()`: One-to-one transformation.
* `filter()`: One-to-zero or one filtering.
* `reduce()`: Many-to-one reduction.

While these functions can be powerful, it's important to note that list comprehensions and generator expressions often provide more readable and sometimes more efficient alternatives for many common use cases, especially for `map` and `filter` operations.  `reduce` is still relevant for certain types of cumulative calculations.

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

````python
def sum_of_even_numbers(numbers):
    """
    Calculates the sum of all even numbers in a list.

    Args:
        numbers: A list of numbers.

    Returns:
        The sum of even numbers in the list. Returns 0 if the list is empty or contains no even numbers.
    """

    if not numbers:  # Check if the list is empty
        return 0

    even_sum = 0
    for number in numbers:
        if number % 2 == 0:  # Check if the number is even
            even_sum += number

    return even_sum


# Example usage:
numbers1 = [1, 2, 3, 4, 5, 6]
result1 = sum_of_even_numbers(numbers1)
print(f"Sum of even numbers in {numbers1}: {result1}")  # Output: 12

numbers2 = [1, 3, 5, 7]
result2 = sum_of_even_numbers(numbers2)
print(f"Sum of even numbers in {numbers2}: {result2}")  # Output: 0

numbers3 = []
result3 = sum_of_even_numbers(numbers3)
print(f"Sum of even numbers in {numbers3}: {result3}")  # Output: 0


numbers4 = [2,4,6,8,10]
result4 = sum_of_even_numbers(numbers4)
print(f"Sum of even numbers in {numbers4}: {result4}") # Output: 30
````

2. Create a Python function that accepts a string and returns the reverse of that string.



````python
def reverse_string(input_string):
    """
    Reverses a given string.

    Args:
        input_string: The string to reverse.

    Returns:
        The reversed string.  Returns an empty string if the input is None or empty.
    """

    if not input_string: # checks for None or empty
      return ""

    return input_string[::-1]  # Efficient string slicing for reversal


# Example Usage
string1 = "hello"
reversed_string1 = reverse_string(string1)
print(f"Original string: {string1}")
print(f"Reversed string: {reversed_string1}")  # Output: olleh

string2 = "Python"
reversed_string2 = reverse_string(string2)
print(f"Original string: {string2}")
print(f"Reversed string: {reversed_string2}")  # Output: nohtyP

string3 = ""
reversed_string3 = reverse_string(string3)
print(f"Original string: {string3}")
print(f"Reversed string: {reversed_string3}")  # Output: (empty string)

string4 = None
reversed_string4 = reverse_string(string4)
print(f"Original string: {string4}")
print(f"Reversed string: {reversed_string4}")  # Output: (empty string)
````

3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.


````python
def square_list(numbers):
    """
    Returns a new list containing the squares of each number in the input list.

    Args:
        numbers: A list of integers.

    Returns:
        A new list containing the squares of the numbers.
        Returns an empty list if the input list is empty or None.
    """

    if not numbers:  # Check if the list is None or empty
        return []

    squared_numbers = []
    for number in numbers:
        squared_numbers.append(number * number)  # Or number**2

    return squared_numbers


# Example Usage:
numbers1 = [1, 2, 3, 4, 5]
squared_list1 = square_list(numbers1)
print(f"Original list: {numbers1}")
print(f"Squared list: {squared_list1}")  # Output: [1, 4, 9, 16, 25]

numbers2 = [-1, 0, 2, -3]
squared_list2 = square_list(numbers2)
print(f"Original list: {numbers2}")
print(f"Squared list: {squared_list2}")  # Output: [1, 0, 4, 9]

numbers3 = []
squared_list3 = square_list(numbers3)
print(f"Original list: {numbers3}")
print(f"Squared list: {squared_list3}")  # Output: []

numbers4 = None
squared_list4 = square_list(numbers4)
print(f"Original list: {numbers4}")
print(f"Squared list: {squared_list4}")  # Output: []
````

4. Write a Python function that checks if a given number is prime or not from 1 to 200.



````python
def is_prime(number):
    """
    Checks if a given number is prime.

    Args:
        number: The number to check for primality.

    Returns:
        True if the number is prime, False otherwise.
        Handles edge cases for numbers less than 2 and non-integer inputs.
    """

    if not isinstance(number, int):  # Check if it's an integer
        return False  # or raise TypeError("Input must be an integer")

    if number < 2:  # Numbers less than 2 are not prime
        return False

    # Optimization: Check divisibility only up to the square root of the number
    for i in range(2, int(number**0.5) + 1):
        if number % i == 0:
            return False  # Divisible, so not prime

    return True  # No divisors found, so it's prime



# Example usage (checking primes from 1 to 200):
for num in range(1, 201):
    if is_prime(num):
        print(f"{num} is prime")

# Example with a specific number:
num_to_check = 97
if is_prime(num_to_check):
    print(f"{num_to_check} is prime")
else:
    print(f"{num_to_check} is not prime")


#Example with non-integer input
print(is_prime(3.14)) #Output: False

print(is_prime("abc")) #Output: False
````

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.


````python
class FibonacciIterator:
    def __init__(self, n):
        """
        Initializes the Fibonacci iterator.

        Args:
            n: The number of Fibonacci terms to generate.
        """
        self.n = n
        self.a = 0
        self.b = 1
        self.count = 0

    def __iter__(self):
        """
        Returns the iterator object itself (required for iterators).
        """
        return self

    def __next__(self):
        """
        Generates the next Fibonacci number.

        Raises:
            StopIteration: If the specified number of terms has been reached.
        """
        if self.count < self.n:
            result = self.a
            self.a, self.b = self.b, self.a + self.b  # Efficiently update a and b
            self.count += 1
            return result
        else:
            raise StopIteration


# Example Usage:

# Generate and print the first 10 Fibonacci numbers
fib_iter = FibonacciIterator(10)
for num in fib_iter:
    print(num)

# Or, use next() explicitly:
fib_iter2 = FibonacciIterator(5)
print(next(fib_iter2))  # Output: 0
print(next(fib_iter2))  # Output: 1
print(next(fib_iter2))  # Output: 1
print(next(fib_iter2))  # Output: 2
print(next(fib_iter2))  # Output: 3
try:
  print(next(fib_iter2)) # Raises StopIteration
except StopIteration:
  print("Iteration Complete")

# Create a list of the first 7 Fibonacci numbers
fib_list = list(FibonacciIterator(7))
print(fib_list)  # Output:


# Example of handling invalid input (non-positive n)
fib_iter_invalid = FibonacciIterator(-5) # Will not produce any output
for num in fib_iter_invalid:
  print(num)

fib_iter_zero = FibonacciIterator(0) # Will not produce any output
for num in fib_iter_zero:
  print(num)
````

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.


````python
def powers_of_two(exponent):
    """
    A generator function that yields powers of 2 up to a given exponent.

    Args:
        exponent: The maximum exponent (inclusive).

    Yields:
        The powers of 2 from 2^0 up to 2^exponent.
    """
    for i in range(exponent + 1):  # Iterate from 0 to exponent (inclusive)
        yield 2**i


# Example Usage:

# Generate and print powers of 2 up to 2^5
for power in powers_of_two(5):
    print(power)  # Output: 1 2 4 8 16 32

# Convert the generated values into a list:
powers_list = list(powers_of_two(8))
print(powers_list)  # Output: [1, 2, 4, 8, 16, 32, 64, 128, 256]

# Using next() explicitly:
generator = powers_of_two(3)
print(next(generator))  # Output: 1
print(next(generator))  # Output: 2
print(next(generator))  # Output: 4
print(next(generator))  # Output: 8
try:
    print(next(generator))  # Raises StopIteration
except StopIteration:
    print("Iteration Complete")

# Handling invalid input:
for power in powers_of_two(-2): # Will not produce any output
  print(power)

for power in powers_of_two(0): # Will produce only one output: 1
  print(power)
````

7. Implement a generator function that reads a file line by line and yields each line as a string.


````python
def read_file_lines(filepath):
    """
    A generator function that reads a file line by line and yields each line as a string.

    Args:
        filepath: The path to the file.

    Yields:
        Each line of the file as a string (including the newline character).
        Yields an empty string if the file is empty.
        Raises FileNotFoundError if the file does not exist.
    """

    try:
        with open(filepath, 'r') as file:
            for line in file:
                yield line
    except FileNotFoundError:
        raise  # Re-raise the FileNotFoundError to be handled by the caller
    except Exception as e: # Catching other potential exceptions like permission errors
        print(f"An error occurred: {e}")
        yield "" # Yield an empty string to indicate an error, or handle it differently


# Example Usage:

# Create a sample file (for demonstration)
with open("my_file.txt", "w") as f:
    f.write("This is the first line.\n")
    f.write("This is the second line.\n")
    f.write("This is the third line.\n")

# Read and print the lines using the generator
try:
  for line in read_file_lines("my_file.txt"):
      print(line, end="")  # end="" prevents extra newlines as lines already have them

  #Demonstrates the FileNotFoundError
  for line in read_file_lines("non_existent_file.txt"):
      print(line)
except FileNotFoundError:
  print("File not found. Please check the file path.")

````

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.


````python
# Function to convert Celsius to Fahrenheit
celsius_to_fahrenheit = lambda c: (c * 9/5) + 32

# List of temperatures in Celsius
celsius_temperatures = [0, 10, 20, 30, 40, 100]

# Convert to Fahrenheit using map()
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

# Output the converted temperatures
print(f"Temperatures in Fahrenheit: {fahrenheit_temperatures}")
````

10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

````python
# Function to check if a character is a vowel
is_not_vowel = lambda char: char.lower() not in 'aeiou'

# Input string
input_string = "Hello, how are you doing today?"

# Remove vowels using filter()
filtered_string = ''.join(filter(is_not_vowel, input_string))

# Output the modified string
print(f"String without vowels: {filtered_string}")
````

Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
order is smaller than 100,00 €.








````python
# Given list of orders
orders = [
    (34587, "Learning Python, Mark Lutz", 4, 40.95),
    (98762, "Programming Python, Mark Lutz", 5, 56.80),
    (77226, "Head First Python, Paul Barry", 3, 32.95),
    (88112, "Einführung in Python3, Bernd Klein", 3, 24.99)
]

# Function to compute total price with the rule applied
def calculate_order(order):
    order_number, _, quantity, price_per_item = order
    total = quantity * price_per_item
    if total < 100:
        total += 10  # Add 10€ if total is less than 100€
    return (order_number, round(total, 2))

# Apply function to all orders
processed_orders = list(map(calculate_order, orders))

# Output result
print(processed_orders)
````
